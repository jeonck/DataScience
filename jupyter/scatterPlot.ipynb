```python
import pandas as pd

import matplotlib.pyplot as plt

# 샘플 데이터 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 27, 22, 32, 29],
    'Score': [85, 90, 88, 92, 95]
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 데이터프레임 출력
print(df)

# 나이와 점수의 관계를 시각화
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Score'], color='blue')
plt.title('Age vs Score')
plt.xlabel('Age')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```
